```julia
using Plots, DelimitedFiles, LaTeXStrings
```

```julia
q_2s = readdlm("q_err_2stage.dat");
p_2s = readdlm("p_err_2stage.dat");

q_3s = readdlm("q_err_3stage.dat");
p_3s = readdlm("p_err_3stage.dat");

q_4s = readdlm("q_err_4stage.dat");
p_4s = readdlm("p_err_4stage.dat");
```

```julia

#f1(x)=0.2*x^1; f2(x)=x^2; 

plt1=plot(xscale=:log10, yscale=:log10,xlabel="log₁₀(h)",ylabel="log₁₀(qErr)",grid=false,framestyle=:box)

plot!(plt1,q_2s[:,1] , q_2s[:,2] , m =:sq,label="2-stage Lobatto")
plot!(plt1,q_3s[:,1] , q_3s[:,2] , m =:circle,label="3-stage Lobatto")
plot!(plt1,q_4s[:,1] , q_4s[:,2] , m =:star5,label="4-stage Lobatto")




plt2=plot(xscale=:log10, yscale=:log10,xlabel="log₁₀(h)", ylabel="log₁₀(pErr)",grid=false,framestyle=:box)

plot!(plt2,p_2s[:,1] , q_2s[:,2] , m =:sq,label="2-stage Lobatto")
plot!(plt2,p_3s[:,1] , p_3s[:,2] , m =:circle,label="3-stage Lobatto")
plot!(plt2,p_4s[:,1] , p_4s[:,2] , m =:star5,label="4-stage Lobatto")

plot(plt1,plt2,layout=(1,2),legend=:topleft)

annotate!([5e-2], [1e-5], text(L"O(h^2)",10))

annotate!([2e-2], [2e-9], text(L"O(h^4)",10))

annotate!([3e-1], [2e-12], text(L"O(h^6)",10))

#savefig("errors.pdf")
```
