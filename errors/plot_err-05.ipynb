```julia
using Plots, DelimitedFiles, LaTeXStrings
```

```julia
sol = readdlm("2s_Lob_sol_05.dat");


q_2s = readdlm("2s_Lob_05.dat");

q_3s = readdlm("3s_Lob_05.dat");

q_4s = readdlm("4s_Lob_05.dat");
```

```julia
plt1=plot(xlabel="t",ylabel="x",grid=false,framestyle=:box,legend=:topleft)
plot!(plt1,sol[:,1] , sol[:,3],label="EXACT")
scatter!(plt1,sol[:,1] , sol[:,2],m=:circle,ms=3,mc=:black,label="2stage Lobatto")
annotate!(plt1,[0.85], [0.1], text(L"\alpha=\frac{1}{4}",10))


plt2=plot(xscale=:log10, yscale=:log10,xlabel="log₁₀(h)",ylabel="log₁₀(qErr)",grid=false,framestyle=:box)

plot!(plt2,q_2s[:,1] , q_2s[:,2] , m =:sq,label="2-stage Lobatto")
plot!(plt2,q_3s[:,1] , q_3s[:,2] , m =:circle,label="3-stage Lobatto")
plot!(plt2,q_4s[:,1] , q_4s[:,2] , m =:star5,label="4-stage Lobatto")



annotate!(plt2,[2e-2], [0.3e-2], text(L"O(h^2)",10))

annotate!(plt2,[1e-2], [1e-8], text(L"O(h^{3.5})",10))

annotate!(plt2,[7e-2], [0.3e-7], text(L"O(h^{3})",10),legend=:bottomright)

annotate!(plt2,[0.7e-2], [1e-2], text(L"\alpha=\frac{1}{4}",10))


plot(plt1,plt2,layout=(1,2))




savefig("error_05.pdf")
```
