```julia
using Plots, DelimitedFiles, LaTeXStrings
```

```julia
energy2s = readdlm("energy_Lobatto2s.dat");
energy3s = readdlm("energy_Lobatto3s.dat");
energy4s = readdlm("energy_Lobatto4s.dat");

exact = readdlm("energy_exact.dat");


rel_err1 = abs.(exact[:,2] - energy2s[:,2])./maximum(exact[:,2]);
rel_err2 = abs.(exact[:,2] - energy3s[:,2])./maximum(exact[:,2]);
```

```julia
using LaTeXStrings

plt1=plot(xlabel=L"T", ylabel="Energy",grid=false,framestyle=:box)

plot!(plt1,exact[:,1] , exact[:,2],label="Exact")

scatter!(plt1,energy2s[:,1], energy2s[:,2],m=:circle,ms=2,mc=:black,label="2-stage Lobatto")

plt2=plot(xlabel=L"\log_{10}(t)", ylabel=L"\log_{10}(E_{\mathrm{err}}(t))",grid=false,framestyle=:box)

plot!(plt2,log10.(exact[:,1]) , log10.(rel_err1),ls=:dash,label="2-stage Lobatto")
plot!(plt2,log10.(exact[:,1]) , log10.(rel_err2),label="3-stage Lobatto")


plot(plt1,plt2,layout=(1,2))




savefig("energy_and_rel_errors.pdf")
```
